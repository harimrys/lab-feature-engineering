**Check for missing values**

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here
spaceship.dropna(how = "any", inplace = True)
```

```python
spaceship.shape
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
#your code here
spaceship["Cabin"] = spaceship["Cabin"].apply(lambda x: x.split("/")[0])
```

```python
spaceship["Cabin"].unique()
```

- Drop PassengerId and Name

```python
#your code here
spaceship.drop(["PassengerId", "Name"], axis=1, inplace=True)
```

```python
spaceship.head()
```

- For non-numerical columns, do dummies.

```python
#your code here
df_dummies = pd.get_dummies(spaceship, columns=["HomePlanet", "Cabin", "Destination"])
df_dummies
```

**Perform Train Test Split**

```python
# Separate features and target
x = df_dummies.drop("Transported", axis=1)
y = df_dummies["Transported"]
```

```python
#your code here
# Split the data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)

from sklearn.preprocessing import StandardScaler
# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#your code here
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn
```

```python
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
pred
```

- Evaluate your model's performance. Comment it

```python
#Para evaluar el modelo
from sklearn.metrics import accuracy_score, classification_report

#Para calcular la precision
precision = accuracy_score(y_test, pred)
print(f"La precision del modelo es: {precision:.2f}")
```

```python
#Muestra un informe de clasificacion
print(classification_report(y_test, pred))
```
